```python
from rdflib import URIRef, BNode, Literal, Namespace,Graph
from rdflib.namespace import FOAF, DCTERMS, XSD, RDF, SDO, RDFS
import pandas as pd 
import kglab
```

Download and Explore Dataset


```python
nutrition_df= pd.read_csv(r'Nutrition.csv',nrows=500, encoding='latin1',on_bad_lines='skip',sep=";",engine='python')
recipes_df = pd.read_csv(r"Recipes.csv",nrows=500, encoding= 'latin1',on_bad_lines='skip',sep=";",engine='python')
restaurants_df = pd.read_csv(r"Restaurants.csv",nrows=500, encoding= 'latin1',on_bad_lines='skip',sep=";",engine="python")
```

Representation ot the Graph

```python
#namespace
EX = Namespace('http://kg-course.io/food-nutrition/')
SCHEMA = Namespace('')
#generate new graph
g=Graph()

#bind prefix to namespace 
g.bind('ex',EX)
g.bind('schema',SDO)
```

Define Entities and Relations

Recipes 

```python
#create an IRI and set properties
for _, row in recipes_df.iterrows():
    recipe_uri = EX[f"recipe/{row['RecipeId']}"]
    g.add((recipe_uri,RDF.type,SDO.Recipe))
    g.add((recipe_uri,SDO.name,Literal(row['Name'])))
    g.add((recipe_uri,SDO.cookTime,Literal(row['CookTime'],datatype=XSD.duration)))
    g.add((recipe_uri,SDO.prepTime,Literal(row['PrepTime'], datatype=XSD.duration)))
    g.add((recipe_uri,SDO.datePublished,Literal(row['DatePublished'], datatype=XSD.dateTime)))
    g.add((recipe_uri,SDO.image,Literal(row['Images'])))
    g.add((recipe_uri,SDO.recipeCategory,Literal(row['RecipeCategory'])))
    g.add((recipe_uri,SDO.keywords,Literal(row['Keywords'])))
    g.add((recipe_uri,EX['ingredientQuantities'],Literal(row['RecipeIngredientQuantities'])))
    g.add((recipe_uri,SDO.recipeIngredient,Literal(row['RecipeIngredientParts'])))
    g.add((recipe_uri,SDO.recipeYield,Literal(row['RecipeServings'],datatype=XSD.int)))
    g.add((recipe_uri,SDO.recipeYield,Literal(row['RecipeYield'])))
    g.add((recipe_uri,SDO.recipeInstructions,Literal(row['RecipeInstructions'])))
  
```

```python
recipe_by_name = {}
for _, r in recipes_df.iterrows():
    uri = EX[f"recipe/{r['RecipeId']}"]
    name = str(r["Name"]).strip().lower()
    recipe_by_name[name] = uri
```

Nutritions

```python
#create an IRI 
def nutrition_uri(i):
    return EX[f"nutrition/{i}"]

#set properties
for i, row in nutrition_df.iterrows():
    n_uri=nutrition_uri(i)
    g.add((n_uri,RDF.type,SDO.NutritionInformation))
    g.add((n_uri,SDO.description,Literal(row['Description'])))
    g.add((n_uri,SDO.recipeCategory,Literal(row['RecipeCategory'])))
    g.add((n_uri, SDO.calories, Literal(row["Calories"], datatype=XSD.float)))
    g.add((n_uri, SDO.fatContent, Literal(row["FatContent"], datatype=XSD.float)))
    g.add((n_uri, SDO.saturatedFatContent, Literal(row["SaturatedFatContent"], datatype=XSD.float)))
    g.add((n_uri, SDO.cholesterolContent, Literal(row["CholesterolContent"], datatype=XSD.float)))
    g.add((n_uri, SDO.sodiumContent, Literal(row["SodiumContent"], datatype=XSD.float)))
    g.add((n_uri, SDO.carbohydrateContent, Literal(row["CarbohydrateContent"], datatype=XSD.float)))
    g.add((n_uri, SDO.fiberContent, Literal(row["FiberContent"], datatype=XSD.float)))
    g.add((n_uri, SDO.sugarContent, Literal(row["SugarContent"], datatype=XSD.float)))
    g.add((n_uri, SDO.proteinContent, Literal(row["ProteinContent"], datatype=XSD.float)))

    recipe_name = str(row["Name"]).strip().lower()
    if recipe_name in recipe_by_name:
        recipe_uri = recipe_by_name[recipe_name]
        g.add((recipe_uri, SDO.nutrition, n_uri))
  
```

Restaurants

```python
#create an IRI and set properties
for _, row in restaurants_df.iterrows():
    rest_uri = EX[f"restaurant/{row['Restaurant ID']}"]
    g.add((rest_uri, RDF.type, SDO.Restaurant))
    g.add((rest_uri, SDO.name, Literal(row["Restaurant Name"])))
    g.add((rest_uri, SDO.addressLocality, Literal(row["City"])))
    g.add((rest_uri, EX["country"], Literal(row["Country"])))
    g.add((rest_uri, SDO.address, Literal(row["Address"])))
    g.add((rest_uri, EX["localityVerbose"], Literal(row["Locality Verbose"])))
    g.add((rest_uri, SDO.longitude, Literal(row["Longitude"], datatype=XSD.float)))
    g.add((rest_uri, SDO.latitude, Literal(row["Latitude"], datatype=XSD.float)))
    g.add((rest_uri, SDO.servesCuisine, Literal(row["Cuisines"])))
    g.add((rest_uri, EX["currency"], Literal(row["Currency"])))
    g.add((rest_uri, EX["hasTableBooking"], Literal(row["Has Table booking"],datatype=XSD.int)))
    g.add((rest_uri, EX["hasOnlineDelivery"], Literal(row["Has Online delivery"],datatype=XSD.int)))
    g.add((rest_uri, EX["isDeliveringNow"], Literal(row["Is delivering now"],datatype=XSD.int)))
    g.add((rest_uri, EX["switchToOrderMenu"], Literal(row["Switch to order menu"],datatype=XSD.int)))
    g.add((rest_uri, EX["priceRange"], Literal(row["Price range"],datatype=XSD.int)))
    g.add((rest_uri, SDO.aggregateRating, Literal(row["Aggregate rating"], datatype=XSD.float)))
    g.add((rest_uri, EX["ratingColor"], Literal(row["Rating color"],datatype=XSD.float)))
    g.add((rest_uri, EX["ratingText"], Literal(row["Rating text"],datatype=XSD.int)))
    g.add((rest_uri, EX["votes"], Literal(row["Votes"], datatype=XSD.int)))
    g.add((rest_uri, EX["numberOfDishesInCuisines"], Literal(row["Nummber of dishes in cuisines"], datatype=XSD.int)))
    g.add((rest_uri, EX["averageCostForTwoUSD"], Literal(row["Average cost of two in USD"], datatype=XSD.decimal)))
```

```python
for _, r_row in recipes_df.iterrows():
    recipe_uri = EX[f"recipe/{r_row['RecipeId']}"]
    recipe_text = (str(r_row["Keywords"]) + " " + str(r_row["RecipeCategory"])).lower()

    for _, rest_row in restaurants_df.iterrows():
        rest_uri = EX[f"restaurant/{rest_row['Restaurant ID']}"]
        cuisines_text = str(rest_row["Cuisines"]).lower()

        if any(tok in cuisines_text for tok in recipe_text.split()):
            g.add((recipe_uri, EX["likelyAvailableAt"], rest_uri))
```
